```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
output_notebook()

url = "https://drive.google.com/uc?export=download&id=1zXIy8T1KICuXtcbzYCqAtXbYX-2ta2Mu"
Data = pd.read_csv(url,header=None)
Data.head(1000)
Data_matrix=Data.values

Data_array = np.linspace(1,1000,1000)
Light_array = np.linspace(1,1000,1000)
for k in range(1000):
  i = Data_matrix[k]
  i = i.item()
  Data_array[k] = i
  if Data_array[k]<35:
    Light_array[k] = 35
  else:
    Light_array[k] = Data_array[k]

p = figure(width = 800, height = 400)
p.line(np.arange(len(Data_matrix)), Data_array, color="red", line_width = 2)
p.line(np.arange(len(Data_matrix)), Light_array, color="green", line_width = 2)
show(p)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
output_notebook()

url = "https://drive.google.com/uc?export=download&id=1zXIy8T1KICuXtcbzYCqAtXbYX-2ta2Mu"
Data = pd.read_csv(url,header=None)
Data.head(1000)
Data_matrix=Data.values

Data_array = np.linspace(1,1000,1000)
Light_array = np.linspace(1,1000,1000)
for k in range(1000):
  i = Data_matrix[k]
  i = i.item()
  Data_array[k] = i
  if Data_array[k]<35:
    Light_array[k] = 1
  else:
    Light_array[k] = 0

fig, ax = plt.subplots(2, 1, figsize = (20, 8))

ax[0].plot(np.arange(len(Data_matrix)), Data_array, 'green')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Light')
ax[0].grid(True)

ax[1].plot(np.arange(len(Data_matrix)), Light_array, 'red')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Lamp')
ax[1].grid(True)

plt.show()
```
